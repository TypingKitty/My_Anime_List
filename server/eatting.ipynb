```python

a = 0
b = 0
c = 0
d = 0

# Define a list of questions with options and corresponding disorders
questions_with_options_and_disorders = [
    {
        "question": "How much more or less do you feel you worry about your weight and body shape than other people your age?",
        "options": [
            "1. I WORRY A LOT LESS THAN OTHER PEOPLE",
            "2. I WORRY A LITTLE LESS THAN OTHER PEOPLE",
            "3. I WORRY ABOUT THE SAME AS OTHER PEOPLE",
            "4. I WORRY A LITTLE MORE THAN OTHER PEOPLE",
            "5. I WORRY A LOT MORE THAN OTHER PEOPLE"
        ],
        "disorders": ['a', 'b', 'c'],
    },
    {
        "question": "How afraid are you of gaining 2kgs?",
        "options": [
            "1. NOT AFRAID OF GAINING",
            "2. SLIGHTLY AFRAID OF GAINING",
            "3. MODERATELY AFRAID OF GAINING",
            "4. VERY AFRAID OF GAINING",
            "5. TERRIFIED OF GAINING"
        ],
        "disorders": ['a', 'b', 'c'],
    },
    {
        "question": "Did you eat until uncomfortably full in the past 3 months?",
        "options": [
            "1. NEVER",
            "2. RARELY",
            "3. SOMETIMES",
            "4. OFTEN",
            "5. ALWAYS"
        ],
        "disorders": ['b', 'c'],
    },
    {
        "question": "Did you eat large amounts of food when not feeling hungry in the past 3 months?",
        "options": [
            "1. NEVER",
            "2. RARELY",
            "3. SOMETIMES",
            "4. OFTEN",
            "5. ALWAYS"
        ],
        "disorders": ['b', 'c'],
    },
    {
        "question": "Do you consume a small amount of food (i.e., less than 1200 calories/day) on a regular basis to influence your shape or weight?",
        "options": ["Yes", "No"],
        "disorders": ['a'],
    },
    {
        "question": "Do you make yourself throw-up/use diuretics or laxatives/exercise excessively to control your weight and shape?",
        "options": ["Yes", "No"],
        "disorders": ['b'],
    },
    {
        "question": "Do you struggle with a lack of interest in eating or food?",
        "options": ["Yes", "No"],
        "disorders": ['d'],
    },
    {
        "question": "Do you avoid certain or many foods because of such features as texture, consistency, temperature, or smell, or have other people suggested this may be the case for you?",
        "options": [
            "1. NEVER",
            "2. RARELY",
            "3. SOMETIMES",
            "4. OFTEN",
            "5. ALWAY"
        ],
        "disorders": ['d'],
    },
    {
        "question": "Do you avoid certain or many foods because of fear of experiencing negative consequences like choking or vomiting, or have other people suggested this may be the case for you?",
        "options": [
            "1. NEVER",
            "2. RARELY",
            "3. SOMETIMES",
            "4. OFTEN",
            "5. ALWAYS"
        ],
        "disorders": ['d'],
    },
]

# Iterate through the questions with options and collect user responses
for question_info in questions_with_options_and_disorders:
    question = question_info["question"]
    options = question_info["options"]
    disorders = question_info["disorders"]
    
    # Display the question and options
    print(question)
    for option in options:
        print(option)
    
    # Take the user's response
    user_response = input("Enter the number corresponding to your response or 'yes'/'no': ")
    
    # Check the response and increment the relevant disorder variables
    if user_response.lower() == 'yes':
        increment = 4
    elif user_response.isdigit() and 1 <= int(user_response) <= 5:
        increment = int(user_response) - 1
    else:
        increment = 0
    
    for disorder in disorders:
        if disorder == 'a':
            a += increment
        elif disorder == 'b':
            b += increment
        elif disorder == 'c':
            c += increment
        elif disorder == 'd':
            d += increment
print("Disorder A (Anorexia Nervosa):", a)
print("Disorder B (Bulimia Nervosa):", b)
print("Disorder C (Binge Eating Disorder):", c)
print("Disorder D (ARFID):", d)
# Calculate the risk level for each disorder variable based on the criteria
risk_a = "no" if a <= 5 else ("risk" if a <= 8 else "yes")
risk_b = "no" if b <= 10 else ("risk" if b <= 15 else "yes")
risk_c = "no" if c <= 8 else ("risk" if c <= 12 else "yes")
risk_d = "no" if d <= 4 else ("risk" if d <= 7 else "yes")

# Print statements indicating the risk level for each type of disorder

if risk_a == "risk":
    print("You are at risk of Anorexia Nervosa disorder.")
elif risk_a == "yes":
    print("You have Anorexia Nervosa disorder.")


if risk_b == "risk":
    print("You are at risk of Bulimia Nervosa disorder.")
elif risk_b == "yes":
    print("You have Bulimia Nervosa disorder.")


if risk_c == "risk":
    print("You are at risk of Binge Eating Disorder.")
elif risk_c == "yes":
    print("You have Binge Eating Disorder.")


if risk_d == "risk":
    print("You are at risk of Avoidant Restrictive Food Intake Disorder (ARFID).")
elif risk_d == "yes":
    print("You have Avoidant Restrictive Food Intake Disorder (ARFID).")

```
